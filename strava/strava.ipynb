```python
import numpy as np
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt

import warnings



warnings.filterwarnings('ignore')

%matplotlib inline
```

```python
df = pd.read_csv('/content/activities.csv')
df.head()
```

```python
df = df[['ID физической активности',
 'Дата тренировки',
 'Название тренировки',
 'Тип активности',
 'Общее время',
 'Расстояние',
 'Время в движении',
 'Дистанция',
 'Макс. скорость',
 'Набор высоты',
 'Высота низменности',
 'Высота подъема',
 'Макс. уклон',
 'Средний угол уклона']]
```

```python
df.columns = ['id',
 'start_date',
 'name',
 'type_activity',
 'total_time',
 'distance_1',
 'time_in_motion',
 'distance',
 'max_speed',
 'alt_gain',
 'min_alt',
 'max_alt',
 'max_slope',
 'avg_slope']
```

```python
df = df[df.type_activity == 'Велосипед']
```

```python
df = df.drop(columns=['type_activity'])
```

```python
df.head()
```

```python
df.isna().mean()
```

```python
df.distance_1 = df.distance_1.apply(lambda x: x.replace(',', '.')).astype('float64')
```

```python
dict_of_month = {'авг.': '08', 'апр.': '04', 'июл.': '07', 'июн.': '06', 'мар.': '03', 'мая': '05', 'окт.': '10', 'сент.': '09'}
```

```python
def month_to_datetime(date):
  for word, replacement in dict_of_month.items():
    date = date.replace(word, replacement)
  return date
```

```python
df.start_date = df.start_date.apply(lambda x:x.replace(' г.,', ''))
```

```python
df.start_date = pd.to_datetime(df.start_date.apply(month_to_datetime), format='%d %m %Y %H:%M:%S')
```
